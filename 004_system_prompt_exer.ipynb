```python
# Load environment variables from .env file
from dotenv import load_dotenv
load_dotenv()
```

```python
# Create an API Client
import anthropic

client = anthropic.Anthropic()
model="claude-sonnet-4-0"
```

```python
def add_user_message(messages, content):
    messages.append({"role": "user", "content": content})

def add_assistan_message(messages, content):
    messages.append({"role": "assistant", "content": content})

def chat(messages, system=None):
    params = {
        "model":model,
        "max_tokens":1024,
        "messages":messages,
    }

    if system:
        params["system"] = system

    message = client.messages.create(
        **params
    )
    return message.content[0].text

```

```python
messages = []

system = """
You are a specialist in Python programming.
You are concise and to the point.
"""
add_user_message(messages, "Write a Python function that checks a string for duplicate characters.")
answer = chat(messages, system=system)
answer
```
