```python
# Load environment variables from .env file
from dotenv import load_dotenv
load_dotenv()
```

```python
# Create an API Client
import anthropic

client = anthropic.Anthropic()
model="claude-sonnet-4-0"
```

```python
def add_user_message(messages, content):
    messages.append({"role": "user", "content": content})

def add_assistan_message(messages, content):
    messages.append({"role": "assistant", "content": content})

def chat(messages, system=None, temperature=1.0):
    params = {
        "model":model,
        "max_tokens":1024,
        "messages":messages,
        "temperature": temperature
    }

    if system:
        params["system"] = system

    message = client.messages.create(
        **params
    )
    return message.content[0].text

```

```python
messages = []

add_user_message(messages, "Muestrame el Preambulo de la Constitucion Argentina")
stream = client.messages.create(
    model=model,
    max_tokens=1024,
    messages=messages,
    stream=True
)

# for event in stream:
#     print(event)

# Imprimir en tiempo real a medida que llegan los mensajes
for event in stream:
    if event.type == "content_block_delta":
        # Este es el texto que va llegando
        print(event.delta.text, end="", flush=True)
    elif event.type == "message_stop":
        print()  # Salto de línea al final
```
